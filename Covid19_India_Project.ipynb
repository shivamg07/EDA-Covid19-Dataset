```python
#dropping redundant columns from the dataframe
covid_df.drop(['Sno','Time','ConfirmedForeignNational','ConfirmedIndianNational'], axis='columns', inplace=True)
```

```python
covid_df.head()
```

```python
# checking datatypes of each of the attributes
covid_df.dtypes
```

```python
# changing datatype of Date column from object to date using to_datetime()
covid_df['Date']=pd.to_datetime(covid_df['Date'], format='%Y-%m-%d')
covid_df.head()
```

```python
#Removing some unwanted rows from the dataframe
covid_df.drop(covid_df[covid_df['State/UnionTerritory'].isin(['Unassigned', 'Cases being reassigned to states'])].index, axis='rows', inplace=True)
```

```python
# checking null/missing values in the dataframe
covid_df.isna().sum()
```

```python
# Renaming some values in column 'State/UnionTerritory'
covid_df.replace({'Maharashtra***': 'Maharashtra', 'Bihar****':'Bihar', 'Karanataka': 'Karnataka','Madhya Pradesh***':'Madhya Pradesh', 'Dadra and Nagar Haveli and Daman and Diu':'Daman & Diu'}, inplace=True)
```

```python
#Find active cases
covid_df['Active_Cases']=covid_df['Confirmed']-covid_df['Cured']-covid_df['Deaths']
covid_df.head()
```

```python
covid_df[covid_df['State/UnionTerritory']=='Kerala']
```

```python
# Pivot table to show Confirmed, Cured, Deaths and Active_Cases Statewise
statewise = covid_df.pivot_table(values=['Confirmed','Cured','Deaths', 'Active_Cases']
                           , index='State/UnionTerritory', aggfunc=max)
```

```python
statewise['Recovery_Rate']= statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise['Mortality_Rate']= statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise = statewise.sort_values(by='Confirmed', ascending=False)
```

```python
statewise.head()
```

```python
statewise.style.background_gradient(cmap='cubehelix')
```

```python
# Top 10 active cases states
top_10_active_cases = covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases']].sort_values(by='Active_Cases',ascending=False).reset_index()

fig = plt.figure(figsize=(12,6))
plt.title('Top 10 States with most active cases in India')
sns.barplot(data=top_10_active_cases.iloc[:10], x='State/UnionTerritory',y='Active_Cases', width=0.5)
plt.xlabel('States')
plt.ylabel('Total Active Cases')
```

```python
# Top 10 states with highest deaths
top_10_states_deaths = covid_df.groupby('State/UnionTerritory').max()[['Deaths']].sort_values('Deaths', ascending=False).reset_index()
plt.figure(figsize=(9,5))
plt.title('Top 10 states with highest Deaths')
sns.barplot(data=top_10_states_deaths[:10], x='State/UnionTerritory', y='Deaths', width=0.5)
plt.xlabel('States')
plt.xticks(rotation=45)
plt.ylabel('Total Deaths')
plt.show()
```

```python
# Growth trend of active cases in top 5 impacted states

fig = plt.figure(figsize=(12,5))
sns.lineplot(covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra', 'Karnataka','Tamil Nadu', 'Uttar Pradesh', 'Kerala'])],
             x='Date', y='Active_Cases', hue='State/UnionTerritory')
plt.title('Active Cases trend of Top 5 States impacted in India')
```

```python
# States with Mortality rate less than 0.5%
plt.figure(figsize=(8,4))
mort_data = statewise[statewise['Mortality_Rate'] <= 0.5].sort_values(by='Mortality_Rate', ascending = True)
ax = sns.barplot(mort_data, y=mort_data.index, x='Mortality_Rate', width = 0.5)
plt.ylabel('State')
plt.xlabel('Mortality Rate %')
plt.title('States with Mortality Rate less than 0.5%')
```

```python
# States with Mortality rate more than 2%
plt.figure(figsize=(9,4))
mort_data = statewise[statewise['Mortality_Rate'] >= 2].sort_values(by='Mortality_Rate', ascending = False)
ax = sns.barplot(mort_data, y=mort_data.index, x='Mortality_Rate', width = 0.5, color='pink')
plt.ylabel('State')
plt.xlabel('Mortality Rate %')
plt.title('States with Mortality Rate more than 2%')

plt.figure(figsize=(10,4))
ax = sns.barplot(statewise[:20], x = statewise[:20].index, y='Confirmed', width = 0.5, color = 'purple')
plt.xlabel('State')
plt.xticks(rotation=90)
plt.ylabel('Confirmed Cases')
plt.gca().get_xticklabels()[0].set_color('red') #get current axis
plt.gca().get_xticklabels()[16].set_color('red')
plt.title('Confirmed Cases in Top 20 States')
```

```python
#dropping columns with high number of null values
vaccine_df.drop(['Sputnik V (Doses Administered)','AEFI','18-44 Years (Doses Administered)','45-60 Years (Doses Administered)','60+ Years (Doses Administered)'], axis='columns', inplace=True)
```

```python
vaccine_df.drop(['Male(Individuals Vaccinated)','Transgender(Individuals Vaccinated)','Female(Individuals Vaccinated)','18-44 Years(Individuals Vaccinated)',
                 '45-60 Years(Individuals Vaccinated)','60+ Years(Individuals Vaccinated)','Total Doses Administered'], axis='columns', inplace=True)
```

```python
vaccine_df.head()
```

```python
#Remove rows where state is India
vaccine = vaccine_df[vaccine_df['State']!='India']
vaccine.head()
```

```python
vaccine.rename(columns={'Total Individuals Vaccinated':'Total'}, inplace=True)
```

```python
# changing the vaccine_Date column to datetime datatype
vaccine['Vaccine_Date']=pd.to_datetime(vaccine['Vaccine_Date'], format='%d/%m/%Y')
vaccine.info()
```

```python
vaccine.head()
```

```python
# Covaxin vs Covishield Doses Administered
covaxin = vaccine_df[' Covaxin (Doses Administered)'].sum()
covishield = vaccine_df['CoviShield (Doses Administered)'].sum()
fig = px.pie(names=['Covaxin','Covishield'], values=[covaxin,covishield], title='Covaxin and Covishield Doses Administered')
fig.update_layout(width = 400, height=400)
```

```python
# Percentage of First and Second doses adminisiterd
first_dose = vaccine_df['First Dose Administered'].sum()
second_dose = vaccine_df['Second Dose Administered'].sum()
fig = px.pie(names=['First_Dose','Second_Dose'], values=[first_dose,second_dose], title='First and Second Doses Administered')
fig.update_layout(width = 400, height=400)

# Out of total individuals vaccinated, 81% got first dose and 19% got second dose
```

```python
#Most vaccinated States
max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total').sort_values('Total', ascending=False)[:5]
plt.figure(figsize=(12,6))
sns.barplot(data=max_vac, x=max_vac.index, y='Total')
plt.xlabel('State')
plt.ylabel('Vaccinations')
plt.title('Top 5 Vaccinated States in India')
```

```python
#Least vaccinated States
max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total').sort_values('Total', ascending=True)[:5]
plt.figure(figsize=(12,6))
sns.barplot(data=max_vac, x=max_vac.index, y='Total')
plt.xlabel('State')
plt.ylabel('Vaccinations')
plt.title('Top 5 Least Vaccinated States in India')
```

```python
# Trend of vaccination and active cases in Maharashtra

plt.figure(figsize = (10,5))
ax1 = sns.lineplot(vaccine[vaccine['State']=='Maharashtra'], x = 'Vaccine_Date', y='Total', color='green')
ax1.set_xlabel('Date')
ax1.set_ylabel('Vaccinated')
ax2 = ax1.twinx()
sns.lineplot(covid_df[covid_df['State/UnionTerritory']=='Maharashtra'], x = 'Date', y='Active_Cases', color ='red')
ax2.set_title('Vaccination vs Active Cases trend of Maharashta')
```
